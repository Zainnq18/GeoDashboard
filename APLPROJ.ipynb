```python
import osmnx as ox
import folium

# Step 1: Download India's road network
india_graph = ox.graph_from_place("India", network_type="drive")

# Step 2: Convert to GeoDataFrames
nodes, edges = ox.graph_to_gdfs(india_graph)

# Step 3: Create a Folium map centered over India
m = folium.Map(location=[22.5, 80], zoom_start=5, tiles="cartodbpositron")

# Step 4: Add road edges as GeoJSON
folium.GeoJson(edges).add_to(m)

# Step 5: Save to HTML file
m.save("india_map.html")
print("Map saved to india_map.html")
```
