```python
import json

# Input and output file paths
geojson_path = "/Users/zain/Downloads/INDIA_STATES-3.geojson"     # Input GeoJSON file
output_js_path = "states_data.js"         # Output JS file

# Load GeoJSON data
with open(geojson_path, 'r', encoding='utf-8') as f:
    geojson_data = json.load(f)

# Write JS file with correctly cased variable name
with open(output_js_path, 'w', encoding='utf-8') as f:
    f.write("var states_data = ")
    json.dump(geojson_data, f)
    f.write(";")  # End statement

print(f"'states_data.js' has been successfully created.")
```

```python
import json
import re
import pandas as pd

# Path to your .js file
js_file_path = "sub_districts_data.js"

# Read and clean the JS content
with open(js_file_path, "r", encoding="utf-8") as f:
    js_content = f.read()

# Remove the JavaScript variable assignment (e.g., var sub_districts_data = ...)
json_str = re.sub(r'^var\s+\w+\s*=\s*', '', js_content).rstrip(';')

# Convert to a Python dictionary
data = json.loads(json_str)

# Extract desired fields
records = []
for feature in data.get("features", []):
    props = feature.get("properties", {})
    records.append({
        "stname": props.get("stname"),
        "dtname": props.get("dtname"),
        "sdtname": props.get("sdtname")
    })

# Create DataFrame
df = pd.DataFrame(records)

# Save to Excel
excel_output_path = "sub_districts_output.xlsx"
df.to_excel(excel_output_path, index=False)

print(f"Excel file saved to: {excel_output_path}")
```

```python
import json
import pandas as pd
import re

# Path to your pincodes_data.js file
js_file_path = 'pincodes_data.js'

# Read the JS file as text
with open(js_file_path, 'r', encoding='utf-8') as f:
    js_text = f.read()

# Extract JSON part inside the JS variable declaration
json_match = re.search(r'var\s+pincodes_data\s*=\s*(\{.*\});', js_text, re.DOTALL)

if not json_match:
    raise ValueError("Could not find pincodes_data JSON object in the JS file.")

json_str = json_match.group(1)

# Parse JSON string to Python dict
data = json.loads(json_str)

# Extract all properties dicts from features
properties_list = [feature['properties'] for feature in data['features']]

# Convert list of dicts to DataFrame
df = pd.DataFrame(properties_list)

# Save DataFrame to Excel file
df.to_excel('pincodes_data.xlsx', index=False)

print("Excel file 'pincodes_data.xlsx' created successfully.")
```
